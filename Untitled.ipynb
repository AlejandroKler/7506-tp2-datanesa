```python
clicks = pd.read_csv("clicks.csv")
```

```python
installs = pd.read_csv("installs.csv")
```

```python
brand_clicks = clicks.groupby("ref_hash").agg({"specs_brand":"mean"})
```

```python
installs["cantidad"] = 1
```

```python
installs_cant = installs.groupby("ref_hash").agg({"cantidad":"sum"})
```

```python
a = installs_cant.merge(brand_clicks, right_index=True, left_index=True, how="left")
```
